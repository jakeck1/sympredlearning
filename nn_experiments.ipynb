### convergence for different parameters

```python
datapath = os.path.join('data','network','convergence')
plotpath = os.path.join('plots','network','convergence')
```

```python
n= 6
n_repetitions=30
MW = np.zeros((n_repetitions, n, n))
MV = np.zeros((n_repetitions, n, n))
parameters = np.linspace(-1, 1, n)

exp_params = {'n_cells_layer_1': 20, 'n_cells_layer_2': 20, 'n_states': 10,'dt':0.1}


for k in range(n_repetitions):
        for i in range(n):
            for j in range(n):
                print(i,j)
                alpha = parameters[i]
                beta = parameters[j]
                
                if (alpha !=0.0) or (beta !=0.0):
                
                    l_W,_ = run_convergence_experiment(alpha_1= alpha,beta_1 = beta,loss_relative_to_first_step = True,only_final_loss=True,n_saves=int(1e5),**exp_params)
                
                    _,l_V = run_convergence_experiment(alpha_2= alpha,beta_2 = beta,loss_relative_to_first_step = True,only_final_loss=True,n_saves=int(1e5),**exp_params)
                
                MW[k,i,j]=l_W
                MV[k,i,j]=l_V
suffices = ['parameter_sweep_w_negative']
save_convergence_data((MW,MV),datapath,suffices=suffices)
```

```python
n = 6
parameters = np.linspace(-1, 1, n)

suffices = ['parameter_sweep_w_negative']

MW,MV = load_convergence_data(datapath,suffices=suffices)
fig = plot_convergence_matrix(MW,parameters,logscale=True,title=r'Convergenge of $W^r$' )
plt.savefig(os.path.join(plotpath,'conv_matrix_W_alpha_beta_w_negative.png'),dpi=300)
fig = plot_convergence_matrix(MV,parameters,logscale=True,title=r'Convergenge of $W^f$')
plt.savefig(os.path.join(plotpath,'conv_matrix_V_alpha_beta_w_negative.png'),dpi=300)


```
