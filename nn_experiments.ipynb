```python

n= 10
n_repetitions=10
parameters = np.linspace(0, 1, n)
```

```python

MW = np.zeros((n_repetitions, n, n))
MV = np.zeros((n_repetitions, n, n))



for k in range(n_repetitions):
        print(k)
        for i in range(n):
            print(i)
            for j in range(n):

                alpha = parameters[i]
                beta = parameters[j]
                
                if (alpha !=0.0) or (beta !=0.0):
                
                    l_W,_ = run_convergence_experiment(alpha_1= alpha,beta_1 = beta,loss_relative_to_first_step = True,only_final_loss=True,n_saves=1e5)
                
                    _,l_V = run_convergence_experiment(alpha_2= alpha,beta_2 = beta,loss_relative_to_first_step = True,only_final_loss=True,n_saves=1e5)
                
                    MW[k,i,j]=l_W
                    MV[k,i,j]=l_V

suffices = ['parameter_sweep']
save_convergence_data((MW,MV),datapath,suffices=suffices,overwrite=True)
```
